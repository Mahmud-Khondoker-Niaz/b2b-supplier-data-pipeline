# Product & Supplier Insights Report

## 1. Introduction
This report presents the findings from an exploratory analysis of product and supplier data from the production database. The goal was to identify data quality issues, extract meaningful insights, and provide actionable business recommendations based on the unified dataset.

---

## 2. Data Overview
The analysis was performed on a dataset comprising four key tables from the integrated catalogue:

| Table | Rows |
| :--- | :--- |
| **Products** | 10,884 |
| **SupplierProductPriceLists** | 14,445 |
| **SupplierProductVariants** | 28,897 |
| **SupplierProductStocks** | 22,748 |

---

## 3. Data Quality Assessment
During the exploration phase, several inconsistencies and quality issues were identified and addressed:

*   **Missing Values:** Columns with more than 50% missing values (e.g., `salesUnitId`, `massUnitId`) were dropped to prevent skewed analysis.
*   **Imputation:** Remaining missing text fields were filled with **'Unknown'**, and numerical fields were defaulted to **0** to maintain data integrity.
*   **Duplicates:** A thorough check confirmed no duplicate records across all four tables.
*   **Outliers:** Statistical analysis flagged **280 products** as high-price outliers (above mean + 3 standard deviations). Additionally, **5 products** were flagged for suspiciously low prices (below 0.10 EUR).

---

## 4. Methodology & Feature Engineering
To prepare the dataset for deeper analysis, five meaningful features were engineered:

| Feature | Description | Business Purpose |
| :--- | :--- | :--- |
| **product_age_days** | Days since creation | Identifies stagnant inventory vs. new arrivals. |
| **price_category** | cheap / mid / expensive | Facilitates market segment analysis. |
| **in_stock** | Boolean (True/False) | Quickly identifies available products for sale. |
| **supplier_prod_count** | Products per supplier | Measures supplier scale and dependency. |
| **category_avg_price** | Avg price per category | benchmarks pricing against category. |

---

## 5. Key Business Insights

### Supplier Performance
*   The supplier landscape is dominated by a few **Large suppliers** (200+ products), who contribute the bulk of the catalogue.
*   Most suppliers fall into the **Small segment** (1-50 products).

### Pricing & Market Position
*   The average product price is **4.88 EUR**, with the majority of the inventory categorized as "Cheap" (under 5 EUR).
*   No currency inconsistencies were found, all pricing is standardized in **EUR**.

### Inventory Status
*   A majority of the products are currently **in stock**, but the flagged outliers suggest potential data entry errors in pricing that could impact sales.

---

## 6. Strategic Recommendations
1.  **Supplier Optimization:** Strengthen partnerships with **Large-segment suppliers** who offer high-volume inventory.
2.  **Pricing Audit:** Manually review the **280 high-price outliers** and the **5 suspiciously low-priced items** to ensure catalog accuracy.
3.  **Data Collection Improvement:** Address the high percentage of missing data in attributes like `brand` and `material` to enhance the shop's filtering capabilities.
4.  **Stock Monitoring:** Implement automated alerts for "Out of Stock" items to minimize lost sales opportunities.

---

## 7. Assumptions & Limitations
*   Due to the absence of direct cost data, **Average Price** was utilized as a proxy for margin-related analysis.
*   The high-price threshold was statistically defined as **mean + 3 standard deviations**.
*   Supplier identification was derived from the `productPropositionCategory` attribute.
